```python
#找出函數的最小值
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def loss(x):
    return x*(x-1)

def de(x):
    return 2*x-1
```

loss function:x*(x-1)
微分:2x-1

```python
x = np.linspace(-5, 5, 1000)
y =  x*(x-1)

plt.plot(x, y);
plt.xlabel("x");
plt.ylabel("y");
```

```python
start = random.uniform(-5,5)
start
```

```python
plt.plot(x, loss(x))
plt.xlabel('x')
plt.ylabel('y')

x=start
learningRate=0.01
for i in range(10000):
    plt.scatter(x, loss(x))
    if loss(x)== -0.25:
        print('min')
        break
    else:
        x -= learningRate*de(x)
```
